#### 2) Print unique values in all columns



```python
Data = pd.read_csv("./Data.csv")
display(Data)


for column in Data.columns:
    print(f"Unique values for column '{column}':")
    display(Data[column].unique())
    print("\n\n")
```

```python
print(Data.isnull().mean() * 100)
```

#### 3) Fil nan value with 'other'

```python
Data = Data.fillna("other")
```

#### 4) Fill NaN in 'agent' with mean of 'agent' column

```python
Data['agent'] = Data['agent'].fillna(Data['agent'].mean())
```

```python
Data.dropna(inplace=True)
```

#### 6) Plot the count of adults and children using a bar plot

#### 7) Perform Label Encoding on categorical columns

```python
Cate_Cols = [cols for cols in Data.columns if cols != "reservation_status_date" and Data[cols].apply(lambda x: isinstance(x, str)).all()]
display(Cate_Cols)

for i in Cate_Cols:
    Encoder = LabelEncoder()
    Data[i] = Encoder.fit_transform(Data[i].values.reshape(-1, 1))

display(Data.head(10))
```
