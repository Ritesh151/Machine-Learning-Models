```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
Data = pd.read_csv("Data.csv")
display(Data)

Cols = Data.columns
print("\nColumns in DataSet = \n", Cols)
```

```python
Information = pd.DataFrame(Data.info())
display(Information)

Basic_Stas = pd.DataFrame(Data.describe())
display(Basic_Stas)
```

### Checking the Null values in data sets

```python
Checking_Null = Data.isnull().mean() * 100 
print("\nNull Values: \n", Checking_Null)
```

```python
# Shortcut for the cleaning for categorical data

string_cols = [col for col in Data.columns if Data[col].dtype == 'object']
print(string_cols)

for i in string_cols:
    Data[i].fillna(Data[i].mode()[0], inplace=True)
```

### Data Cleaning for Price

```python
Price = Data['Price']
display(Price)

Info = pd.DataFrame(Price.info())
display(Info)

Data['Price'] = Data['Price'].fillna(Data['Price'].median())
Null_Price= Data['Price'].isnull().mean() * 100
print("\nNull Values = ", Null_Price)
```

### Data Cleaning for Distance

```python
Distance = Data["Distance"]
display(Distance)

Null = Distance.isnull().sum()
print("\nNull = ", Null)

'''
Info = Distance.info()
print("\nInformation of Distance", Info)
'''

Data['Distance'] = Data['Distance'].fillna(Data['Distance'].median())
Null = Distance.isnull().sum()
print("\nChecking Again Null Values = ", Null)
```

### Data Cleaning for Postcode

```python
Postcode = Data['Postcode']
display(Postcode)

Null = Postcode.isnull().sum()
print("\nNull Values in Postcode = ", Null)

'''
Info = Postcode.info()
print("\nInformation = ", Info)
'''

Data['Postcode'] = Data['Postcode'].fillna(Data['Postcode'].median())
Null = Postcode.isnull().sum()
print("\nChecking Again Null Values = ", Null)
```

### Shortcut for Data Cleaning by Ritesh Gajjar

```python
import warnings
warnings.filterwarnings('ignore')

Numeric_Cols = [col for col in Cols if (Data[col].dtype == "float64" or Data[col].dtype == "int64")]
# print(Numeric_Cols)
Categorical_Cols = [col for col in Cols if (Data[col].dtype == "object")]
# print(Categorical_Cols)

for col in Numeric_Cols:
    Data[col].fillna(Data[col].median(), inplace=True)

for col in Categorical_Cols:
    Data[col].fillna("No Data Found", inplace=True)

Null = Data.isnull().mean() * 100
print(Null)
```

### Basic Visualization

```python
sns.pairplot(Data, kind="hist")
```

```python
plt.figure(figsize=(15,30))
plt.title("Property Distribution")
cols = ['Suburb', 'Address', 'Rooms', 'Type', 'Method', 'SellerG', 'Distance', 'Postcode', 'Bedroom2', 'Bathroom', 'Car', 'Landsize', 'BuildingArea', 'YearBuilt', 'CouncilArea', 'Lattitude', 'Longtitude', 'Regionname', 'Propertycount']
for i in range(len(cols)):
    plt.subplot(5,3,i+1)
    plt.title(cols[i])
    plt.bar(x=Data[cols[i]], height=Data['Price'])
plt.tight_layout()
plt.show()
```
