# Part-1: data Exploration and Pre-processing



```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

#### 1) Load the given dataset

#### 2) Print all the column names

#### 3) Describe the data

```python
Data = pd.read_csv("./Data.csv")
display(Data)

Cols = Data.columns
print("\nColumns Are = \n", Cols.to_list())

print("\n\nData Statistics")
Stat = pd.DataFrame(Data.describe())
display(Stat)
```

#### 4) Drop address, date, postcode, YearBuilt, lattitude, longtitude columns

```python
Data = Data.drop(["Address", "Postcode", "YearBuilt", "Longtitude", "Lattitude"], axis=1)
```

#### 5) Find the count of null value in each column

#### 6) Fill the null value of property count, distance, Bedroom2, Bathroom, Car with 0

#### 7) Fill Null value of land size and bidding area columns with Mean

```python
Total_Val = len(Data.index)
print("Total Values = ", Total_Val)


# 5) Find the count of null value in each column
Sum_Null_Val = Data.isnull().sum()
print("Counting the null values for each column = \n", Sum_Null_Val)



# 6) Fill the null value of property count, distance, Bedroom2, Bathroom, Car with 0
Data['Propertycount'] = Data['Propertycount'].fillna(0)

Data['Distance'] = Data['Distance'].fillna(0)

Data['Bedroom2'] = Data['Bedroom2'].fillna(0)

Data['Bathroom'] = Data['Bathroom'].fillna(0)

Data['Car'] = Data['Car'].fillna(0)


# 7) Fill Null value of land size and Building area columns with Mean
Data['Landsize'] = Data['Landsize'].fillna(Data['Landsize'].mean())
Data['BuildingArea'] = Data['BuildingArea'].fillna(Data['BuildingArea'].mean())
```

#### 8) Find the unique value in method column

#### 9) Create a dummy data for categorical data

```python
# 8) Find the unique value in method column
Method = Data['Method'].unique().tolist()
print("\nUnique Values in Method Coumns = ", Method)


# 9) Create a dummy data for categorical data
Categorical_cols = [col for col in Data.columns if Data[col].dtype == 'object']
print(list(Categorical_cols))

DD = pd.get_dummies(Categorical_cols)
display(DD)
```

```python
Data['Price'] = Data['Price'].fillna(Data['Price'].median())

Export = Data.to_csv("Part_1.csv", index=False)
```
